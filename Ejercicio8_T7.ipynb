Mínimos cuadrados para realizar la proyección del vector b=(-3,-3,8,9) en el subespacio 
dado por los vectores: u_1=(3,1,0,1), u_2=(1,2,1,1), u_3=(-1,0,2,-1).

```python
A = np.array([[3,1,-1],[1,2,0],[0,1,2],[1,1,-1]])
bb = np.array([[-3],[-3],[8],[9]]) 
print(A)
#print(bb)
AT = np.dot(A.T,A)
bbT = A.T @ bb

x = np.linalg.solve(AT,bbT)
print(x)

P=A@x
P
```

```python
#Gram-schmidt (generalizado)
def grands2(A):
    a=A.T
    B=np.zeros(np.shape(a))
        
    for n in range(np.shape(a)[0]):
        B[n]=a[n]
        for i in range(n):
            B[n]-=(a[n]@B[i])/(B[i]@B[i])*B[i]
        B[n]=B[n]*(1/np.linalg.norm(B[n]))
    return B.T
```

```python
O=grands2(A)
O
```
